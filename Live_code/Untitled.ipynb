```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats as sp 
```

```python
x=np.linspace(0,2*np.pi,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
mean = 0  
std_dev = 0.1
noise = sp.norm.rvs(loc=mean, scale=std_dev, size=y.shape)
y_ns=y+noise
plt.plot(x,y_ns)
```

```python
n=10
result_std=np.zeros(n)

plt.figure(figsize=(12, 10), layout='constrained')
plt.title(r'   ')
plt.xlabel(r'x')
plt.ylabel(r'y')
plt.grid(which='minor', linewidth=0.3, linestyle='--')
plt.minorticks_on()
plt.grid()
plt.plot(x,y_ns,linewidth=1.5,label=' ')

for i in range(n):
   coefficient=np.polyfit(x,y_ns,i+1)
   approx_y=np.polyval(coefficient, x)
   plt.plot(x,approx_y,linestyle='--',linewidth=0.5,label=f'approx for n={i+1}') 
   result_std[i]=np.sqrt(np.sum((y-approx_y)**2)/(y.shape[0]*(y.shape[0]-1)))

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6), layout='constrained')
plt.title(r'    ')
plt.xlabel(r'n')
plt.ylabel(r' , %')
plt.grid(which='minor', linewidth=0.3, linestyle='--')
plt.minorticks_on()
plt.grid()
plt.plot(np.linspace(1,n,n),result_std*100,linewidth=1.5)
plt.show()
```
